````python
# Cleaning
dirty = pd.read_csv('dirty_data.csv')
dirty.shape
````

````python
dirty.head()
````

````python
dirty.isnull().sum()
````

````python
# Handling Missing Values
# 1. Imputation => fillna() => mean, meadian, mode
# 2. Removal    => dropna()
````

````python
# filling the salary column with mean strategy

dirty['salary'].fillna(dirty['salary'].mean(), inplace=True)
````

````python
dirty.isnull().sum()
````

````python
# filling the age column with median strategy

dirty['age'].fillna(dirty['age'].median(), inplace=True)
````

````python
dirty.isnull().sum()
````

````python
# filling the city column with mode strategy


dirty['city'].fillna(dirty['city'].mode()[0], inplace=True)
````

````python
dirty.isnull().sum()
````

````python
dirty.dropna(subset=['name'], inplace=True)
````

````python
dirty.isnull().sum()
````

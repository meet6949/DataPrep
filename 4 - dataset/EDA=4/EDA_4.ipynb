# univariate analysis

```python
df["Amount"].mean()
```

```python
df["Amount"].median()
```

```python
df["Amount"].mode()
```

```python
df["Amount"].min(), df["Amount"].max()
```

```python
df.var(numeric_only=True)
df.std(numeric_only=True)
df["Amount"].skew()
```

```python
fig, axcs=plt.subplots(1, 2, figsize=(15, 5))

sns.histplot(
    df["Amount"],
    kde=True,
    ax=axcs[0])

axcs[0].set_title("Amount Distribution")

sns.boxplot(
    x=df["Amount"],
    ax=axcs[1])

axcs[1].set_title("Amount Outliers")

plt.show()
```

```python
df["Amount"].skew()
df["Amount"].kurt()
```

```python
df["Class"].value_counts()
df["Class"].nunique()
```

```python
fig, ax = plt.subplots(1,2, figsize= (15,5))

sns.countplot(
    x=df["Class"],
    ax=ax[0]
)

df["Class"].value_counts().plot(
    kind="pie",
    autopct="%1.1f%%",
    ax=ax[1]
)

plt.show()
```

# bivariate analysis

### numarical vs numarical

```python
df[["Time","Amount"]].corr()
```

```python
fig, ax = plt.subplots(1,2, figsize=(15,5))

sns.scatterplot(
    x=df["Time"],
    y=df["Amount"],
    ax=ax[0]
)

sns.regplot(
    x=df["Time"],
    y=df["Amount"],
    ax=ax[1]
)

plt.show()
```

### numarical vs catagorical

```python
df.groupby("Class")["Amount"].mean()
```

```python
fig, ax = plt.subplots(1,3, figsize=(15,5))

sns.violinplot(
    x=df["Class"],
    y=df["Amount"],
    ax=ax[0])

sns.barplot(
    x=df["Class"],
    y=df["Amount"],
    ax=ax[1])

sns.boxplot(
    x=df["Class"],
    y=df["Amount"],
    ax=ax[2])

plt.show()
```

### multivariate

```python
corr = df.corr(numeric_only=True)
```

```python
plt.figure(figsize=(10,8))

sns.heatmap(
    corr,
    annot=False
)

plt.show()
```

```python
sns.pairplot(df[["Time","Amount","Class"]])
plt.show()
```

```python
sns.scatterplot(
    x=df["Time"],
    y=df["Amount"],
    hue=df["Class"],
    size=df["Amount"]
)

plt.show()
```

```python
g = sns.FacetGrid(df, col="Class")
g.map(sns.histplot, "Amount")
plt.show()
```
